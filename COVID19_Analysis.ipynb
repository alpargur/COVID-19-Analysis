## Task 2
- Markov Chain (a stochastic model describing a sequence of possible events in which the probability of each event depends only on the state attained in the previous event.)
- **Discrete Time Finite State** : Time is divided into discrete time steps. The system starts at 0 which is the initial time step, the system goes from its current state to the next one chosen with some randomness in a finite space.
- **Time Homogeneous** : The given transition probabilities will be the same irrespective of time. 
- **Probability of Transition** : The probability of going from current state to another state.
- **Transition Matrix** : For a Markov Chain at a given time it holds the information on the probabilities of transitioning between states. Each row is a probability vector and sums up to 1.

[Markov Chain Example](https://www.youtube.com/watch?v=tYaW-1kzTZI&list=PLIeGtxpvyG-KxMqY8wJ0KoY3ujxf0C856&index=24)

[Transition Matrix](https://www.youtube.com/watch?v=tYaW-1kzTZI&list=PLIeGtxpvyG-KxMqY8wJ0KoY3ujxf0C856&index=24)

* **Start Vector** : We initialize start vector to find the [transition probabilities](https://www.youtube.com/watch?v=1_uPFU_49d0&list=PLIeGtxpvyG-KxMqY8wJ0KoY3ujxf0C856&index=26) by multiplying it with the transition matrix. The sum the numbers in start vector equals to 1.

- **Recurrent** : State i is recurrent if it’s expected to come return. When state is recurrent, there is always a way to come back. 

- **Transient** : A state is known as recurrent or transient depending upon whether or not the Markov Chain will eventually return to it. If it’s not recurrent it is transient, which means it’s not returning to that state again. 

- **Absorbing** : Absorbing state is the state with the probability of $Pi,i = 1$. Which means the state is self-transitioning.

- **Irreducible** : A Markov Chain is irreducible if there exists a chain of steps between any two states that has positive probability.

- **Periodicity** : After random state changes with a certain interval if the system is ending on the same state then it is periodic. If there is a self-transition it is aperiodic.

- **Stationary Distribution** : Irreducibility and periodicity both concern the locations a Markov Chain could be at some later point. Stationary Distributions deal with the likelihood of process being in a certain state at an unknown point of time. For a chain with finite number of states, each of which positive recurrent an aperiodic chain is the same as an irreducible chain.

## SIR Model

* It’s a model which is used to simulate the spread of epidemic in order to estimate future scenarios and its effects on a population. There are three states **Susceptible**, **Infected** and **Recovered**. Initial state of the population is constant which means there are no deaths or births in populations and the physical, genetical differences are neglected. 

* The goal is to observe how the spread of the virus changes and how fast it spreads in different populations and under different circumstances so that we can acquire knowledge about properties of each parameter and their impact on the spread of the virus.

* With the implementation of *Markov Chain* in SIR Model, we are predicting the infectiousness of the virus in **discrete time steps**.

* Clustering is an **unsupervised machine learning** algorithm. It helps clustering data points into groups. It is tricky to evaluate clustering results since they don’t contain **ground truth labels** (the ideal expected result).

### K-Means :
* K indicates the number of the variables. If we were to split our model into two, K would be equal to two. This approach is mainly used for classification, to group similar datas.

    ### How to determine the K-value?

    * **Domain Knowledge** : It might give some prior knowledge. It gives more relevant insights. 
    * **Data Driven Approach** : This method is used if domain knowledge is unavailable. The most popular one is Elbow Method.
    * **Clustering Quality** : Ideal clustering is characterized by minimal intra cluster distance and maximal inter cluster distance.

    ### Silhoutte Analysis


* Silhoutte Analysis is used to study the separation distance between the resulting clusters. It gives an insight about the accuracy of the cluster. It’s done by observing the **Silhoutte Coefficient**.
* +1: The sample is far away from the neighbour.
* 0 : Sample location is very close to decision boundary.
* -1: Sample might have been assigned to wrong cluster.

##  Neural Networks
* We have sets of inputs and a set of target values, we're trying to get predictions that match those target values as closely as possible.
    * An input layer x
    * Arbitrary amount of hidden layers
    * An output layer ŷ
    * A set of weights(w) & biases(b) between each layer w,b.
    * An activation function(σ) for each hidden layer.  
        * σ maps any real value into another value between 0 and 1. We use it to map predictions into probability.
        * Activation Function is what distinguishes Neural Network from Linear Regression models. Without activation functions, we have many different combinations of inputs.
    * The output ŷ of a simple 2-layer Neural Network   $ŷ = σ(w_2σ(w_1x + b1) + b2)$
        * Every neuron has its own weights & biases.

    
* Right values for weight & biases determines the **strenght of predictions**. Fine tuning these known as training the Neural Network.
    * Calculating the predicted output ŷ (feedforward)
    * Updating weights & biases (backpropagation)

* **Loss Function** : It evaluates the goodness of our predictions(sum of square errors). The difference is squared so that we measure the **absolute value** of the difference.
* The goal is to find the best set of weights & biases that minimizes the loss.
* **Backpropagation** : After the error measurement we need to update or weights & biases, to do that we need to know the **derivative** of loss function with respect to the weights & biases.
    * To get derivate we apply chain rule for weights & biases.
    * The gradient of slope tells the direction we need to go move towards **minima**. The weights at minima of function is what we want.

## Clustering
- K-Means works nicely when we already have the number of K and according to the experts it's not very good at identifying classes when dealing with groups which don't have spherical distribution shape (since it's calculating the Eucledian distannces between data). But what if we don't know how many classes we have?

- **Hierarchical Clustering** : We don't need to specify the number of clusters, it'll find it by itself.
    - **Agglomerative** : This approach first assumes that each sample is a different cluster then it starts to merge the closer ones until there is only one cluster.
        - **Single Linkage** : Assumes each point is a cluster, then computes the distance between most similar ones then merges.
        - **Complete Linkage** : Unlike Single Linkage, it compares the most dissimilar points of a cluster to perform the merge operation.
    - **Divise** : Englobes all datapoints into one and then splits the cluster iteratively into smaller pieces.
    - **Pros** :
        - The results can be very informative.
        - It especially powerful when the dataset contains hierarchical relationship.
    - **Cons** :
        - It is computationally very expensive.
- **Gaussian Mixture Models (GMM)**
    - A probabilistic model that assumes all samples are generated from a mix of a finite number of Gaussian distributions with unknown parameters. (I find this one really interesting among all clustering methods because it has an extraordinary approach and it benefits from different areas of mathematics.)
    It belongs to **soft clustering** every data point belongs to every cluster existing in dataset with different levels of membership. It's assigned as probability from 0 to 1.
***
- Clustering is widely used in lots of business areas such as movie recommedations or bank mortgages to identify users, customers and decide in which group they belong. My questions is how can we benefit from clustering in our use case? Because epidemic has a dynamic where the number of each group in SIR Model is constantly changing, we also shouldn't neglect the fact that these changes occur from the interactions within a network.


#### In conclusion I happen to came up with this idea:
   If we want to simulate the spread of an epidemic we can level up the quality of our model by implementing a **network** init. We can take a sample group and we would have wide range of people(nodes in this case) with different attributes, such as age, gender, height, fitness level, work, transportation preference, number of people they live with...
  
  While creating a network we would also need to connect our nodes with links and if we have weighted links it would increase the quality of our model.
    
   In our network model, we have bunch of variables such as the properties of our little community, the parameters we have from our SIR Model (infectious rate, mortality rate, recovery speed) and we can tweak those parameters to see what kind of an influence they make on our model. In that way would have a living, more dynamic model.
